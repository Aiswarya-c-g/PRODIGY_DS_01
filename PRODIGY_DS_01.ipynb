```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
file_path = "population.xlsx"  # Make sure this file is in your working directory
excel_data = pd.ExcelFile(file_path)

# Load data from the correct sheet
df = excel_data.parse('Sheet1')

# Clean the dataset
data = df.iloc[3:, :]  # Skip metadata rows
data.columns = df.iloc[2]  # Use actual column headers
data = data[1:]  # Remove the row that was promoted to header

# Convert population values in 2024 to numeric, ignore non-numeric values
data['2024.0'] = pd.to_numeric(data[2024.0], errors='coerce')

# Drop rows with missing 2024 population data
data = data.dropna(subset=['2024.0'])

# Plot a histogram
plt.figure(figsize=(10, 6))
plt.hist(data['2024.0'] / 1e6, bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Population (in millions) in 2024')
plt.xlabel('Population (millions)')
plt.ylabel('Number of Countries')
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.show()
```
